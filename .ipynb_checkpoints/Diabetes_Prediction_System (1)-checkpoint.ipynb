```python
features=df.drop(columns='diabetes',axis=1)
# Capping outliers
def rem_outliers(df,columns):
    q1=df[columns].quantile(0.25)
    q3=df[columns].quantile(0.75)
    IQR=q3-q1
    lb=q1-1.5*IQR
    ub=q3+1.5*IQR
    # df=df[(df[columns]>lb)&(df[columns]<ub)]
    df[columns]=df[columns].apply(lambda a : lb if a<lb else ub if a>ub else a)
    return df
for column in features:
    df=rem_outliers(df,column)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
log_regression=LogisticRegression()
log_regression.fit(x_train1,y_train)
```

```python
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

# Evaluating the algorithms

```python
from sklearn import metrics
```

```python
score1=metrics.f1_score(y_test,y_predict1)
score2=metrics.f1_score(y_test,y_predict2)
score3=metrics.f1_score(y_test,y_predict3)
score4=metrics.f1_score(y_test,y_predict4)
score5=metrics.f1_score(y_test,y_predict5)
```

```python
print(score1,score2,score3,score4,score5)
```

```python
mse1=metrics.mean_squared_error(y_test,y_predict1)
mse2=metrics.mean_squared_error(y_test,y_predict2)
mse3=metrics.mean_squared_error(y_test,y_predict3)
mse4=metrics.mean_squared_error(y_test,y_predict4)
mse5=metrics.mean_squared_error(y_test,y_predict5)
```

```python
print(mse1,mse2,mse3,mse4,mse5)
```

```python
from sklearn.metrics import classification_report
```

```python
print("Using Logistic Regression \n",classification_report(y_test,y_predict1),"\n\n")
print("Using Support Vector Classifier \n",classification_report(y_test,y_predict2),"\n\n")
print("Using Gradient Booster Classifier \n",classification_report(y_test,y_predict3),"\n\n")
print("Using Random Forest Classifier \n",classification_report(y_test,y_predict4),"\n\n")
print("Using KNN \n",classification_report(y_test,y_predict5),"\n\n")
```
