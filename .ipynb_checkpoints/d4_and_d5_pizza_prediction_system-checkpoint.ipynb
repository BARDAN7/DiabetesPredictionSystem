```python
# Find the categorical columns
categorical_col=[]
continous_col=[]
for col in df.columns:
    if df[col].nunique()<=10:
        categorical_col.append(col)
    else:
        continous_col.append(col)
print("Categorical Columns is:\n",categorical_col)
print("Continous Columns is:\n",continous_col)
```

```python
from scipy.stats import zscore
```

```python
categorical_col.extend(['topping','variant'])
```

```python
for x in ['variant','topping']:
    continous_col.remove(x)
continous_col
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_x=LabelEncoder()
for lab in categorical_col:
    df.loc[:,lab] =label_x.fit_transform(df.loc[:,lab])
df[categorical_col]
```

```python
# Inter Quartile Range
def remove_outliers(df,columns):
    for x in columns_to_detect:
        q1= df[x].quantile(0.25)
        q3= df[x].quantile(0.75)
        IQR=q3-q1
        lb=q1-1.5*IQR
        ub=q3+1.5*IQR
        df[columns]=df[columns].apply(lambda x: lb if x<lb else x).copy()
        df[columns]=df[columns].apply(lambda x: ub if x>ub else x).copy()
        return df
for columns in columns_to_detect:
    df=remove_outliers(df,columns).copy()
    
```

```python
sns.boxplot(data=df[categorical_col])
plt.xticks(rotation=30)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns={'price'},axis=1)
y=df['price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# model training
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
svr=SVR()
svr.fit(x_train,y_train)
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
```

```python
y_predict1=lr.predict(x_test)
```

```python
y_predict2=svr.predict(x_test)
```

```python
y_predict3=rf.predict(x_test)
```

```python
y_predict4=gb.predict(x_test)
```
